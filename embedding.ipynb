```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
# sentences 
sent=[ 'the glass of milk',
       'the glass of juice',
       'the cup of tea',
       'I am a good boy',
       'I am a good developer',
       'Understand the meaning of words',
       'The cup of coffee']

sent_1=[
    'The cup of good developer of words'
]
```

```python
sent
```

```python
## Define the vocabulary size
voc_size=10000
```

```python
### One Hot Representation
one_hot_repr=[one_hot(words,voc_size) for words in sent]
one_hot_repr
```

```python
### One Hot Representation
one_hot_repr_1=[one_hot(words,voc_size) for words in sent_1]
one_hot_repr_1
```

```python
## word embedding Representation

from tensorflow.keras.layers import Embedding
from tensorflow.keras.utils import pad_sequences 
# pad_sequences is used to reshape the data(sentence) such that all the sentences have same length -- this is required because the input to the RNN should have fixed size
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

```python
sent_length=8
embedded_docs=pad_sequences(one_hot_repr,padding='pre',maxlen=sent_length)
print(embedded_docs)

```

```python
# featuree representation
dim=10
## dimensions means feature space used to represent the words because each word is represented by a vector of size dim
```

```python
model = Sequential()
model.add(Embedding(voc_size, dim))
model.compile(optimizer='adam', loss='mse')
model.build(input_shape=(None, sent_length))  # Build the model
model.summary()
```

```python
model.predict(embedded_docs)
```

```python

print(embedded_docs[0])
print('-------------------------------------------')

# Get prediction for 1st sentence
pred = model.predict(embedded_docs[0:1])

print(f"Shape: {pred.shape}")           # (1, 8, 10)
print(f"\nFirst word embedding:\n{pred[0][0]}")   # 10 values for 1st word
```
